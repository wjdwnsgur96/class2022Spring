### Constructing a DataFrame using dict [Dict로 df(데이터프레임) 만들기!]

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df 

#여기에서 dict는 다섯개가 정의가 되어 있다. 
#각 item에 대해서 colon(:)을 기준으로 구분할 수 있다. (왼쪽에 있는건 key, 오른쪽은 value)
#key는 열의 카테고리, value는 행에 입력된 데이터값으로 생각하면 된다.
#왼쪽(key)은 str의 형태, 오른쪽(value)은 list의 형태
#pd.dataframe(data) 명령어를 통해 위에 적어놓은 key와 value를 데이터프레임화 시킬수 있다. 
```

### Constructing a DataFrame using pd.Series

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df

#column에 해당하는 내용을 입력하지 않고 데이터 값만 입력하여 panda의 series 함수를 통해 df 생성 가능.
#내용(value)값만 입력하면 됨.
#Key는 sepal을 앞에 붙여서 정의만 해주면 됨.
```

### Changing column names

```python
df.columns
```

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df

#df의 columns들을 리스트화해서 바꾸고 싶은 column값을 재정의할 수 있다.
#여기서는 'SPECIES'를 'species'로 바꿔서 입력함. 
#결과적으로 column의 'SPECIES'가 'species'로 바뀌어서 df에 나타남.
```

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']

#개별 column을 수정하고 싶을 때, 개별 column에 access하는 과정.
#loc=타겟 column을 특정하는 명령어
#(:)은 '전부 다'를 지칭하는 명령어, 뒤에 있는 colon index(length)의 전부 다를 가져오라는 명령어
#=모든 row에서 'sepal_length'에 해당하는 column만 가져와라 
```

```python
# Access one column (short-hand)
df['sepal_length']

#아니면 그냥 df하고 바로 'sepal length' 입력하면 됨. 
```

```python
# Access one column (shortest-hand)
df.sepal_length

#가장 간편한 방법
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]

#두개의 columns를 가져오고 싶을 때
#그냥 두개 다 쓰면 됨
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]

#하나의 Row 정보를 가져오는 방법
#마찬가지로 df.loc를 사용
#여기에선 (:)을 앞이 아닌 뒤에 넣는다.
#=Row 3에서 모든 column을 가져와라
```

```python
# Access two rows
df.loc[[3, 0], :]

#두개의 Row를 가져오는 방법
#마찬가지로 Row 대괄호에 더 찾고 싶은 Row 추가
```

```python
# Access a range of rows
df.loc[2:5, :]

#2,3,4 Row를 전체 가져와라
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]

#'sepal_length'가 5보다 큰 경우만 가져와라
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4]

#df.loc을 이용
#=모든 row에서 'sepal_length' column의 4번 row에 있는 데이터를 선별하여 가져와라
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
df

#새로운 column을 추가하는 법, (=) 다음에 새 column에 들어갈 데이터값 입력
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column
df

#column을 빼는 방법
#axis=1(열축)에서 'new column'을 찾아서 없애라 
```

### Adding/Removing new rows

```python
df.loc[5, :] = ''
df

#Row를 추가하는 방법
#Row5를 만들어서 모든 열에 ''(비어있는 값)을 넣어라
```

```python
df = df.drop(5, axis=0) # axis=0: row
df

#Row를 제거하는 방법
#df에 axis=0(행축)에 있는 Row 5를 드랍하라
```
