### Constructing a DataFrame using dict [Dict로 df(데이터프레임) 만들기!]

````python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df 

#여기에서 dict는 다섯개가 정의가 되어 있다. 
#각 item에 대해서 colon(:)을 기준으로 구분할 수 있다. (왼쪽에 있는건 key, 오른쪽은 value)
#key는 열의 카테고리, value는 행에 입력된 데이터값으로 생각하면 된다.
#왼쪽(key)은 str의 형태, 오른쪽(value)은 list의 형태
#pd.dataframe(data) 명령어를 통해 위에 적어놓은 key와 value를 데이터프레임화 시킬수 있다. 
````

### Constructing a DataFrame using pd.Series

````python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df

#column에 해당하는 내용을 입력하지 않고 데이터 값만 입력하여 panda의 series 함수를 통해 df 생성 가능.
#내용(value)값만 입력하면 됨.
#Key는 sepal을 앞에 붙여서 정의만 해주면 됨.
````

````python
items = r['response']['body']['items']
items[0]

#item이 무엇이 들어있는지 보는 명령어
#여기서 item은 list의 형태로 나옴 (대괄호로 구성)
#강남구(list의 첫번째 아이템)만 보고 싶다면 items[0]을 입력
````

````python
items = r['response']['body']['items']

cityName = [item['cityName'] for item in items]
pm10Value = [item['pm10Value'] for item in items]
pm25Value = [item['pm25Value'] for item in items]
so2Value = [item['so2Value'] for item in items]
coValue = [item['coValue'] for item in items]
o3Value = [item['o3Value'] for item in items]
no2Value = [item['no2Value'] for item in items]

cityName = pd.Series(cityName)
pm10Value = pd.Series(pm10Value)   
pm25Value = pd.Series(pm25Value)
so2Value = pd.Series(so2Value)
coValue = pd.Series(coValue)
o3Value = pd.Series(o3Value)
no2Value = pd.Series(no2Value)

df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})

#pd.DataFrame에 속해 있는 구성 요소들은 df의 Column에 해당함
#pd.Series(panda의 Series 함수)를 이용하여 데이터프레임(df) 생성 가능
````

````python
def test(sidoName):
  URL = ('http://apis.data.go.kr/B552584/ArpltnStatsSvc/'
      'getCtprvnMesureSidoLIst?'
      'sidoName='+sidoName+
      '&searchCondition=hour'
      '&pageNo=1'
      '&numOfRows=100'
      '&returnType=json'
      '&serviceKey=miH%2BZXg85lQy4%2FkmhffvygXDIFiTwisriSLxtyECSzw6hxTVK7yI8QKIOc7mP7LEqCnhBGuj6iW1gZW65dum7A%3D%3D')
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  cityName = [item['cityName'] for item in items] #list comprehension 이용
                                                  #for loop를 통하여 25번 돌아감, 'cityname' 중 강남구에 해당하는 것들 선별
                                                  #마우스를 list의 items에 올려놓으면 항목 갯수 등 확인 가능
  pm10Value = [item['pm10Value'] for item in items]
  pm25Value = [item['pm25Value'] for item in items]
  so2Value = [item['so2Value'] for item in items]
  coValue = [item['coValue'] for item in items]
  o3Value = [item['o3Value'] for item in items]
  no2Value = [item['no2Value'] for item in items]

  cityName = pd.Series(cityName)
  pm10Value = pd.Series(pm10Value)   
  pm25Value = pd.Series(pm25Value)
  so2Value = pd.Series(so2Value)
  coValue = pd.Series(coValue)
  o3Value = pd.Series(o3Value)
  no2Value = pd.Series(no2Value)

  df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})
  return df
````
